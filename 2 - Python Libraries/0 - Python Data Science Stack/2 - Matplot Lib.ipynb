<p>Matplot Lib is used to creates plots and graphs for Data Visualization.</p>

----

<p>- Plots: it's used to check out how the datas are distributed.</p>
<p>- Graphics: it's used to show the analysis and processing results.</p>

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
matplotlib.__version__
```

<h2 style="color:blue">0 - Plots</h2>

```python
x = [1, 2, 3, 4, 5]
y = [1, 3, 7, 4, 5]

plt.plot(x, y, label=['Just Only Datas'])

plt.xlabel('Random X Values')
plt.ylabel('Random Y Values')
plt.title('Random Title')

plt.legend()
plt.show()
```

----

<h2 style="color:blue">1 - Graphics</h2>

```python
x = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
y = [1, 3, 7, 4, 5]

plt.bar(x, y, color=['red', 'blue', 'green', 'yellow', 'orange'])

plt.xlabel('Months')
plt.ylabel('Profits')
plt.title('Profit per Month (per thousand dolars)')

plt.show()
```

```python
x = [1, 2, 3, 4, 5, 2, 3, 1, 2, 5, 5, 5, 4, 1.5, 2.5, 3.5, 4.5, 2.75]

plt.hist(x, histtype='stepfilled')
plt.show()
```

```python
x = [1, 2, 3.5, 1.25, 2.5]
y = [1, 3, 3, 2, 2.5]

plt.scatter(x, y, color='green', marker='x')
plt.show()
```

```python
labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
y = [1, 3, 7, 4, 5]

plt.pie(y, labels=labels, colors=['red', 'blue', 'green', 'yellow', 'orange'], shadow=True)
plt.show()
```

```python
y = [1, 2, 3, 4, 5]
x1 = [1, 3, 7, 4, 5]
x2 = [1, 2, 3, 4, 10]

plt.stackplot(y, x1, x2, colors=['green', 'red'])
plt.show()
```
